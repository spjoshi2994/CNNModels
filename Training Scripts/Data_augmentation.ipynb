```python
import cv2
import numpy as np
from skimage import io
from glob import glob
import random
from scipy.misc import imresize
```

```python
def FillImg(binImg):
    _, contours, hierarchy = cv2.findContours(np.uint8(binImg), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    newBinImg = np.zeros((binImg.shape[0],binImg.shape[1]),np.uint8())
    areas = [cv2.contourArea(c) for c in contours]
    max_index = np.argmax(areas)
    cnt = contours[max_index]
    cv2.drawContours(np.uint8(newBinImg), [cnt], 0, 255, -1)
    return newBinImg

def mobile_data_creation(mobile,background,height,width):
    ''' width1 = int((width)/4)
    length1 = int((height)/2)'''
    img = cv2.resize(mobile,(height,width),cv2.INTER_AREA)
    x_axis = random.randint(1,64)
    y_axis = random.randint(1,64)
    binMask = np.zeros((height*2,width*2)).astype('uint8')
    zeros = np.argwhere(img<250)
    zeros = zeros.transpose([1,0])
    zeros = zeros[0:2]
    zeros = zeros.transpose([1,0])
    zeros = np.unique(zeros,axis=0)
    bgImg = background
    bgImgResized = cv2.resize(bgImg,(width*2,height*2),cv2.INTER_AREA)
    for i in range(len(zeros)):
        bgImgResized[zeros[i][0]+x_axis][zeros[i][1]+y_axis][:]=img[zeros[i][0]][zeros[i][1]][:]
        binMask[zeros[i][0]+x_axis][zeros[i][1]+y_axis] = 255
    binMask = FillImg(binMask)
    return bgImgResized,binMask    
```

```python
bgImgPath = '/home/ubuntu/DeepMobile/data/colour_data/background_Machine/355355084334365_bottom_1000.png'
mob = cv2.imread('/home/ubuntu/DeepMobile/data/colour_data/mobiles/front/355355084334365_back_12.png')
bgImg = cv2.imread(bgImgPath)
x,y = mobile_data_creation(mob,background=bgImg,height=256,width=324)
```
